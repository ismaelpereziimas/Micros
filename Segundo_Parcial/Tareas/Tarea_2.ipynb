**FILTRO DE QUEEEE???**

Un filtro de media móvil simplemente reemplaza cada punto de la señal por el promedio de ese punto y sus vecinos más cercanos. Esto "diluye" los picos de ruido repentinos y revela la forma principal de la señal.


Una de sus fórmulas (porque se puede implementar de distintas maneras) es la mostrada en la imagen:

image.png

donde:

M= es cuantos datos vamos a considerar para el promedio. (Ojito que para esta fórmula tiene que ser par).  
n= es el índice del arreglo que estás evaluando.   
k= este valor es el que permite "apuntar" al dato que se agregará a la suma





# ***Veamos un ejemplo*** ✅

Aunque se ve muy imponente, la realidad es que es muy sencilla la fórmula.  
Veamos un ejemplo:  

Imaginemos que tenemos el array A=[3, 4, 5, 6, 7, 8, 9, 4].



Aplicaremos el filtro de medias móviles y almacenaremos los valores en otro array que llamaré B. Para que podamos aplicar el promedio necesitamos que el número que estamos evaluando tenga un vecino a la izquierda y uno a la derecha.  



1. Análisis de A[0]

  Analicemos elemento por elemento. El primero es el elemento A[0], es decir el 3. Este no tiene vecinos el vecino izquierdo, por lo que lo copiamos tal cual en el array B,el cual ya tiene su primer elemento:

                                B= [3]

2. Análisis de A[1]

A[1] SI TIENE VECINOS DERECHO E IZQUIERDO, por lo que si podemos calcular el promedio, sumamos a sus vecinos con él, es decir,
                
                     A[0]+A[1]+A[2] = 3+4+5= 12

y como son tres elementos, para sacar el promedio lo dividimos entre 3.

                                 12/3 = 4
al obtener el resultado lo añadimos a la lista B, la cuál queda de la siguiente forma:

                                 B=[3,4]


3. Análisis de A[2]

A[2] SI TIENE VECINOS DERECHO E IZQUIERDO, por lo que si podemos calcular el promedio, sumamos a sus vecinos con él, es decir,
                
                      A[1]+A[2]+A[3] = 4+5+6= 15

y como son tres elementos, para sacar el promedio lo dividimos entre 3.

                                15/3 = 5
al obtener el resultado lo añadimos a la lista B, la cuál queda de la siguiente forma:

                                B=[3,4,5]

De esta forma repito el procedimiento hasta haber terminado con todos los elementos del array. El array B[] corresponde a la señal filtrada.
